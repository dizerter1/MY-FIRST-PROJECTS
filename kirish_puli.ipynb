```python
url = "https://assets.datacamp.com/production/repositories/2210/datasets/794e0759b73a2d80baa5d8fb88636a47965139d3/college_datav3.csv"
```

# Bu yerda menga  kerakli bolgan data basani urli kochirib olingan

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df= pd.read_csv(url)
df.head()
```

# Tepada ozimga kerakli kutubhonalarni chaqirib oldim va urlini olgan kutubxonamni dataframe korinishiga keltirib oldim va head orqali bazzi joylarini koz orqali tekshirdim

```python
sns.set_style('white')
sns.displot(df['Tuition'])
```

## Bu esa displot orqali tuitoin yani oqish narxi desak boladi oshani count qilingani default holatda

```python
sns.set(color_codes=True)
sns.displot(df['Tuition'],color='k')
```

## bu yerda yana displotni ichidagi color funksiyasi orqali uning rangini ozgartirib olidk

```python
fig,ax = plt.subplots()
sns.histplot(df['Tuition'],ax=ax)
ax.set(xlabel="Tuition",ylabel="count",xlim=(0,50000),title="Tuition of stats university")
```

## Tepada set orqali uning x va y labellariga nom kiritdik va tituliga ham nima ekanini yozib qoydik eng asosiy yangi ishimiz bu xlim orqali ozimizga keraak bolgan qismini ajratib oldik yani 0 dan 50000 gacha

```python
fig, (ax0,ax1) = plt.subplots(nrows=1,ncols=2,sharey=True,figsize=(9,5))

sns.histplot(df['Tuition'],stat='density',ax=ax0)

sns.histplot(df.query('REGION== 5')['Tuition'],stat='density',ax=ax1)

ax1.set(xlabel='Tuition (5)',xlim=(0,70000))
ax1.axvline(x=20000,label="MY BUDGET",linestyle='--')
ax1.legend()
```
