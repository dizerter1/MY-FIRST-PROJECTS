```python
url = "https://assets.datacamp.com/production/repositories/2210/datasets/fb4f2c1039e3df2c2e2624a8c95de5a1980861c6/bike_share.csv"
```

## kerakli url olingan

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv(url)
```

## kutubxonalarnichaqirib oldik

```python
df
```

## malumotlarni tekshirish  uchun ularni shunchaki yozib kordik

```python
sns.regplot(data=df, x='temp',y='total_rentals'
, marker='+')
```

## regplot orqali chizma chizdik va uni x va y oqlarini temp va total_rentals bilan oldik markerini + shaklida oldik . bu yerda chiziq bu bizga malumotlar qanday rivojlanganligini korsatib beradi yani osgan holatda

```python
sns.residplot(data=df, x='temp',y='total_rentals')
```

## x va y ozgarmagan holatda ularni residplotga berdik. u bizga undan oldingi ozgarishni anglatgan kodni shunchaki tahlilga osonlashtirish uchun gorizontal korinishga keltirdi

```python
sns.regplot(data=df, x='temp',y='total_rentals'
, marker='+',order=2)
```

## regplotga order 2 korinishni berganimizda u bizga chiziqni malumotlarni medianasidan otqazdi desak ham boladi, bu bizga ancha qulayliklar olib keladi. Masalan malumotlar qachon kotarilib qachon tushayotganini bilish imkonini beradi 

```python
sns.residplot(data=df, x='temp',y='total_rentals'
,order=2)
```

## order 2 korinishini residplotga berganimizda u bizga chiziqni malumotlarni medianasidan otkazib uni togri shaklga keltirdi

```python
sns.regplot(data=df, x='mnth',y='total_rentals'
,x_jitter=.1,order=2)
```

## biz regplotni x_jitter funksiyasiga .1 qiymatini berganimizdan song u bizga oylarda bikelarni sotib olinishini shunday korinishda olib berdi , endi bizga aniq korinib turibdiki qaysi oyda hajm oshgan va tushgan

```python
sns.regplot(data=df, x='mnth',y='total_rentals'
, x_estimator=np.mean,order=2)
```

## biz regplotga x_estimatoriga np.mean ni berganimizdan song bizga har oyda biklarni olinishini ortachasini shunday korsatib beradi

```python
sns.regplot(data=df, x='temp',y='total_rentals'
, x_bins=4,order=2)
```

## x_binsda esa u templarni 4 guruhga teng bolib korsatib berdi

```python
df_crosstab =  pd.crosstab(df["mnth"], df["weekday"],values=df["total_rentals"],aggfunc="mean").round(0)
df_crosstab
```

# biz df_crosstab orqali datadagi bikelar olinishini oylarning hafta kunlarini ortachasini korsatib beruvchi jadval hosil qildik

```python

sns.heatmap(pd.crosstab(df["mnth"], df["weekday"],
values=df["total_rentals"], aggfunc="mean"))
```

## heatmap bu bizga juda kerakli narsa bolib unda biz hozir tepadagi jadvalni aniqroq tushunishimiz uchun ranglar korinishida bergan , bu jadvalda rang qanchalik och bolsa hajm shunchalik katta

```python
sns.heatmap(df_crosstab, annot=True, fmt=".1f",cmap="YlGnBu", cbar=True,linewidths=.1)
```

## bu yerda esa tepadagi jadvalni ozimizga kerakli rangda korishimiz mumkin, annot funksiyasi sonlarni korsatish uchun,cmap ozimizga kerakli rang,cbar esa ong tomondagi ranglarni ozgarish shkalasi uchun

```python
sns.heatmap(df_crosstab, annot=True, fmt=".1f",cmap="YlGnBu", cbar=True,center=df_crosstab.loc[1,0])
```
