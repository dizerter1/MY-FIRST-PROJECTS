```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

# Bu yerda menga  kerakli kutubxonalar chaqirilgan


```python
yuung_people_survey = pd.read_csv("young-people-survey-responses.csv")
yuung_people_survey.head()
```

## tepada kerakli csv faylini data frame ga oqib olib head bilan tekshirdim

```python
k = yuung_people_survey["Internet usage"].value_counts()
k
```

## shunchaki qiymatlarni sanadim

```python
# fig,ax = plt.subplots()
sns.catplot(x="Internet usage",y="Age",data=yuung_people_survey,hue="Gender",kind="point")
plt.xticks(rotation=15)
plt.show()
```

## Internet usage qiymatlarini Age ga solishtirganman . bu yerdan shuni korish mumkinki eng katta chetlanish 'less than an hour a day' da korinib turubdi. Togri bundan kattarogini 'no time at all' korish mumkin lekin bu qiymatlar kamligi uchun shunday bolgan

```python
sns.set_style("ticks")
sns.catplot(x="Internet usage",y="Age",data=yuung_people_survey,hue="Gender",kind="point")
plt.xticks(rotation=15)
plt.show()
```

## bu yerda ozgargan yagona narsa bu set_style orqali qilingan ozgarish hisoblanadi

```python
sns.set_style("dark")
sns.catplot(x="Internet usage",y="Age",data=yuung_people_survey,hue="Gender",kind="point")
plt.xticks(rotation=15)
plt.show()
```

## Bu yerda set_styleni dark qilib ishlatilgan

```python
sns.set_style("darkgrid")
sns.set_palette("RdBu")
sns.catplot(x="Internet usage",y="Age",data=yuung_people_survey,hue="Gender",kind="point")
plt.xticks(rotation=15)
plt.show()
```

## Bu yerda set_styleni darkgrid qilib ishlatilgan va set_palette ni ishlatib rang tanlaganmiz


```python
sns.catplot(x="Internet usage",data=yuung_people_survey,hue="Internet usage",kind="count")
plt.xticks(rotation=15)

plt.show()
```

## hozir catplot orqali ustunlarni va korinishida boldi buyerda x Internet usage qiymatlari qoyilgan va y ga ularni count qilgan

```python
sns.set_palette("Greys")
sns.catplot(x="Internet usage",data=yuung_people_survey,hue="Internet usage",kind="count")
plt.show()
```

## SHunchaki rangi ozgartirilgan

```python
sns.set_style("darkgrid")
sns.set_palette("RdBu")
sns.set_context("talk")
sns.catplot(x="Internet usage",y="Age",data=yuung_people_survey,hue="Gender",kind="point")
plt.xticks(rotation=90)

plt.show()
```

## set_context orqali korinishini talk rejimiga otkazdik

```python
sns.set_style("darkgrid")
sns.set_palette("RdBu")
sns.set_context("notebook")
sns.catplot(x="Internet usage",y="Age",data=yuung_people_survey,hue="Gender",kind="point")

plt.show()
```

## bu yerda esa notebook ga 

```python
sns.set_style("darkgrid")
sns.set_palette("RdBu")
sns.set_context("paper")
sns.catplot(x="Internet usage",y="Age",data=yuung_people_survey,hue="Gender",kind="point")

plt.show()
```

## bu yerda paperga

```python
# sns.set_style("darkgrid")
sns.set_palette("RdBu")
sns.set_context("paper")
sns.catplot(x="Internet usage",y="Age",data=yuung_people_survey,hue="Gender",kind="point")
plt.xticks(rotation=15)
plt.grid()
plt.show()
```

## matplotlibni ozini grid funksiyasi bor ekan
