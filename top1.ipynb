```python

medals = pd.read_csv("medals_by_country_2016.csv",index_col=0)
fix , ax = plt.subplots()
ax.bar(medals.index,medals["Gold"])
ax.set_xticklabels(medals.index, rotation=90)

plt.show()
```

```python
medals.head()
```

```python
fix , ax = plt.subplots()
ax.bar(medals.index,medals["Gold"])
ax.bar(medals.index,medals["Silver"],bottom=medals["Gold"])

ax.set_xticklabels(medals.index, rotation=90)
plt.show()
```

```python
fix , ax = plt.subplots()
ax.bar(medals.index,medals["Gold"])
ax.bar(medals.index,medals["Silver"],bottom=medals["Gold"])
ax.bar(medals.index,medals["Bronze"],bottom=medals["Gold"]+medals["Silver"])
ax.set_xticklabels(medals.index, rotation=90)
plt.show()
```

```python
fix , ax = plt.subplots()
ax.bar(medals.index,medals["Gold"],label = "Gold")
ax.bar(medals.index,medals["Silver"],bottom=medals["Gold"],label = "Silver")
ax.bar(medals.index,medals["Bronze"],bottom=medals["Gold"]+medals["Silver"],label = "Bronze")
ax.set_xticklabels(medals.index, rotation=90)
ax.legend()
plt.show()
```

```python

mens_Rowing = summer2016[(summer2016["Sex"] == "M") & (summer2016["Sport"] == "Rowing")]
mens_gymnastics = summer2016[(summer2016["Sex"] == "M") & (summer2016["Sport"] == "Gymnastics")]
# mens_Rowing.head()
mens_gymnastics.head()
```

```python
fig , ax = plt.subplots()
ax.bar("Rowing",mens_Rowing["Height"].mean(),color = 'b')
ax.bar("Gymnastics",mens_gymnastics["Height"].mean(),color = "r")
ax.set_ylabel("Height (sm)")
plt.show()
```

```python
fig , ax = plt.subplots()
ax.hist(mens_Rowing["Height"])
ax.hist(mens_gymnastics["Height"])
ax.set_xlabel("Height (cm)")
ax.set_ylabel("# of observations")
plt.show()
```

```python
fig , ax = plt.subplots()
ax.hist(mens_Rowing["Height"],label="Rowing")
ax.hist(mens_gymnastics["Height"],label="gymnastics")
ax.set_xlabel("Height (cm)")
ax.set_ylabel("# of observations")
ax.legend()
plt.show()
```

```python
fig , ax = plt.subplots()
ax.hist(mens_Rowing["Height"],label="Rowing",bins=5)
ax.hist(mens_gymnastics["Height"],label="gymnastics",bins=5)
ax.set_xlabel("Height (cm)")
ax.set_ylabel("# of observations")
ax.legend()
plt.show()
```

```python
fig , ax = plt.subplots()
ax.hist(mens_Rowing["Height"],label="Rowing",bins=[150,160,170,180,190,200,210],histtype="step")
ax.hist(mens_gymnastics["Height"],label="gymnastics",bins=[150,160,170,180,190,200,210],histtype="step")
ax.set_xlabel("Height (cm)")
ax.set_ylabel("# of observations")
ax.legend()
plt.show()
```
