```python
import pandas as pd
import numpy as np
import seaborn as sns

df = pd.read_csv('goldsilver.csv')
```

## kerakli kutubxonalr va data

```python
df
```

## datani tekshirish

```python
df = df.drop(columns='Element')
```

## str bolgan ustunni tashlab yubordik chunki standardisation faqat numeric qiymatlarga ishlaydi

```python
sns.histplot(x='BoilingPoint',data=df)
```

## BoilingPoint ustunining hisplotda korinishi standardisationdan oldingi

```python
x_score_max_scaling = df / df.max()
x_score_max_scaling
```

## standardisationning max scaling usulidan keyingi ozgarish

```python
sns.histplot(x='BoilingPoint',data=x_score_max_scaling)
```

## uni ozgarishini grafikda korishingiz mumkin, ozgarish shuki ortacha qiymat ortada bolmaydi, limit -1 va 1 oraligida boladi yani max va min

```python
x_score_mean_scaling = (df-df.mean())/(df.max()-df.min())
x_score_mean_scaling
```

## bu esa mean scaling uslubida qiymatlarni ozgarishi

```python
sns.histplot(x='BoilingPoint',data=x_score_mean_scaling)
```

## grafiklarda uni korishingiz mumkin,ozgarish shuki ortasi va ortachsi 0 boladi, elementlar -1 va 1 oraligida yotadi

```python
x_score_minmax_scaling = (df-df.min())/(df.max()-df.min())
x_score_minmax_scaling
```

## bu esa minmax scalingdan keyingi ozgarishlar

```python
sns.histplot(x='BoilingPoint',data=x_score_minmax_scaling)
```

## grafikda shunday boladi, bunda elementlar 0 va 1 orasida yotadi

```python
x_score_robust_scaling = (df-df.median())/(df.quantile(.75)-df.quantile(0.25))
x_score_robust_scaling
```

## bu robust scalingdan keyingi holat

```python
sns.histplot(x='BoilingPoint',data=x_score_robust_scaling)
```

## grafik holatda bu ham -1 va 1 oraligida yotadi ammo korib turganingizdek bu har doim ham aniq emas ulardan katta yo kichik bollish ehtimoli ham bor

```python
x_score_standardisation = (df-df.mean())/df.std()
x_score_standardisation
```

## bu esa standardisationdan keyingi holat


```python
sns.histplot(x='BoilingPoint',data=x_score_standardisation)
```
